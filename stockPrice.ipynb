#이동평균
- 추이를 확인하고 싶을때

```python
ma_df = df.copy()
ma_df.head()
```

```python
ma_days = [30, 70, 90, 120, 150]
for day in ma_days:
  column_name = f"ma_{day}days"
  ma_df[column_name] = ma_df["close"].rolling(day).mean()
ma_df.head()
```

```python
fig_list = ["close", "ma_30days", "ma_70days", "ma_90days", "ma_120days", "ma_150days"]

fig = go.Figure()
for i in range(len(fig_list)):
  fig.add_trace(go.Line(x=ma_df.index, y=ma_df[fig_list[i]], name=fig_list[i]))
fig.update_layout(title="ma graph", width=1000, height=500)
fig.show()
```

```python
data = df.filter(["close"])
print(data.shape)

dataset = data.values
print(dataset.shape)

#95%데이터만 학습에 사용
training_data_len = int(np.ceil(len(dataset)*.95)) #1674
print(f"학습에 사용할 데이터 개수: {training_data_len}")

#scaling MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
scaled_data = mms.fit_transform(dataset)
print(f"scale한 데이터 shape: {scaled_data.shape}")

#학습데이터 분리
train_data = scaled_data[:1674, :]
print(f"학습데이터로 사용할 train data shape: {train_data.shape}")


# x_train = [[1, 2, 3, 4, 5, 6],
#            [2, 3, 4, 5, 6, 7],
#            [3, 4, 5, 6, 7, 8]]

# y_train = [1, 2, 4, 4, 6, 7]


x_train = []
y_train = []

for i in range(60, len(train_data)): # i =  60 ~ 1674  >>>>>>> 60일을 1 사이클  전 시점이 현재 시점에 영향을 주는 모델이 lstm 0~60, 1~61, 2~62, 3~63
  x_train.append(train_data[i-60:i, 0])
  y_train.append(train_data[i, 0])

print(f"x_train length: {len(x_train)}")
print(f"y_train length: {len(y_train)}")
# print(y_train[:5])

x_train = np.array(x_train)
y_train = np.array(y_train)
print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")

x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
print(f"lstm모델에 넣기 위한 x_train shape: {x_train.shape}")


##model 생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

model = Sequential()
model.add(LSTM(128, input_shape=(x_train.shape[1], 1), return_sequences=True))
model.add(LSTM(64, activation="relu"))
model.add(Dense(25))
model.add(Dense(1))

print(model.summary())

#5%데이터만 테스트에 사용
```
